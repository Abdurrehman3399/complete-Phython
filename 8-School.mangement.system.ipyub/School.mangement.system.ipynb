```python
from openpyxl import Workbook, load_workbook
import os

# ================== PARENT CLASS ==================
class Person:
    def __init__(self, name="", age="", gender=""):
        self.name = name
        self.age = age
        self.gender = gender


# ================== STUDENT CLASS ==================
class Student(Person):
    def __init__(self):
        super().__init__()

    def add_student(self):
        self.name = input("Enter Name: ")
        self.age = input("Enter Age: ")
        self.gender = input("Enter Gender: ")
        self.roll_no = input("Enter Roll Number: ")
        self.class_name = input("Enter Class Name: ")
        self.marks = input("Enter Marks: ")

    def calculate_grade(self):
        marks = int(self.marks)
        if marks >= 80: return "A"
        elif marks >= 60: return "B"
        elif marks >= 40: return "C"
        else: return "Fail"

    def save_to_excel(self):
        file = "students_data.xlsx"

        if not os.path.exists(file):
            wb = Workbook()
            wb.active.append(["Name","Age","Gender","Roll No","Class","Marks","Grade"])
            wb.save(file)

        wb = load_workbook(file)
        sheet = wb.active
        sheet.append([self.name,self.age,self.gender,self.roll_no,self.class_name,self.marks,self.calculate_grade()])
        wb.save(file)
        print("✅ Student saved!")

    def show_students(self):
        file = "students_data.xlsx"
        if not os.path.exists(file):
            print("No data!")
            return

        wb = load_workbook(file)
        for row in wb.active.iter_rows(values_only=True):
            print(row)

    def delete_student(self):
        file = "students_data.xlsx"
        roll = input("Enter Roll No: ")

        wb = load_workbook(file)
        sheet = wb.active

        for i in range(2, sheet.max_row+1):
            if str(sheet.cell(i,4).value) == roll:
                sheet.delete_rows(i)
                wb.save(file)
                print("Deleted")
                return
        print("Not found")


# ================== TEACHER CLASS ==================
class Teacher(Person):
    def __init__(self):
        super().__init__()

    def add_teacher(self):
        self.name = input("Enter Name: ")
        self.age = input("Enter Age: ")
        self.gender = input("Enter Gender: ")
        self.id = input("Enter Teacher ID: ")
        self.subject = input("Enter Subject: ")
        self.salary = input("Enter Salary: ")

    def save_teacher(self):
        file = "teachers_data.xlsx"

        if not os.path.exists(file):
            wb = Workbook()
            wb.active.append(["Name","Age","Gender","ID","Subject","Salary"])
            wb.save(file)

        wb = load_workbook(file)
        wb.active.append([self.name,self.age,self.gender,self.id,self.subject,self.salary])
        wb.save(file)
        print("✅ Teacher saved!")

    def show_teachers(self):
        file = "teachers_data.xlsx"
        if not os.path.exists(file):
            print("No data!")
            return

        wb = load_workbook(file)
        for row in wb.active.iter_rows(values_only=True):
            print(row)

    def delete_teacher(self):
        file = "teachers_data.xlsx"
        tid = input("Enter ID: ")

        wb = load_workbook(file)
        sheet = wb.active

        for i in range(2, sheet.max_row+1):
            if str(sheet.cell(i,4).value) == tid:
                sheet.delete_rows(i)
                wb.save(file)
                print("Deleted")
                return
        print("Not found")


# ================== COURSE CLASS ==================
class Course:
    file = "courses.xlsx"

    @staticmethod
    def init():
        if not os.path.exists(Course.file):
            wb = Workbook()
            wb.active.append(["Course Name","Course Code","Teacher"])
            wb.save(Course.file)

    @staticmethod
    def add_course():
        Course.init()
        name = input("Course Name: ")
        code = input("Course Code: ")
        teacher = input("Teacher Name: ")

        wb = load_workbook(Course.file)
        wb.active.append([name,code,teacher])
        wb.save(Course.file)
        print("✅ Course added")

    @staticmethod
    def show_courses():
        Course.init()
        wb = load_workbook(Course.file)
        for row in wb.active.iter_rows(values_only=True):
            print(row)

    @staticmethod
    def delete_course():
        Course.init()
        code = input("Enter Course Code: ")

        wb = load_workbook(Course.file)
        sheet = wb.active

        for i in range(2, sheet.max_row+1):
            if str(sheet.cell(i,2).value) == code:
                sheet.delete_rows(i)
                wb.save(Course.file)
                print("Deleted")
                return
        print("Not found")


# ================== MAIN MENU ==================
student = Student()
teacher = Teacher()

while True:
    print("\n===== MAIN MENU =====")
    print("1. Add Student")
    print("2. Show Students")
    print("3. Delete Student")
    print("4. Add Teacher")
    print("5. Show Teachers")
    print("6. Delete Teacher")
    print("7. Add Course")
    print("8. Show Courses")
    print("9. Delete Course")
    print("10. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        student.add_student(); student.save_to_excel()
    elif choice == "2":
        student.show_students()
    elif choice == "3":
        student.delete_student()
    elif choice == "4":
        teacher.add_teacher(); teacher.save_teacher()
    elif choice == "5":
        teacher.show_teachers()
    elif choice == "6":
        teacher.delete_teacher()
    elif choice == "7":
        Course.add_course()
    elif choice == "8":
        Course.show_courses()
    elif choice == "9":
        Course.delete_course()
    elif choice == "10":
        break
    else:
        print("Invalid choice")
```

```python
import tkinter as tk
from tkinter import ttk, messagebox
from openpyxl import Workbook, load_workbook
import os

# ================= CONFIG =================
ADMIN_USERNAME = "branded"
ADMIN_PASSWORD = "29710"

BG = "#ecf0f1"
SIDEBAR = "#2c3e50"
BTN = "#34495e"
ACCENT = "#1abc9c"

# ================= WINDOW =================
root = tk.Tk()
root.title("BrandeD Technologies")
root.geometry("1000x600")
root.config(bg=BG)

# ================= HELPERS =================
def clear_root():
    for widget in root.winfo_children():
        widget.destroy()

def clear_frame(frame):
    for widget in frame.winfo_children():
        widget.destroy()

def create_excel(file, headers):
    if not os.path.exists(file):
        wb = Workbook()
        ws = wb.active
        ws.append(headers)
        wb.save(file)

def calculate_grade(marks):
    try:
        marks = int(marks)
    except:
        return "Invalid"

    if marks >= 80: return "A"
    elif marks >= 60: return "B"
    elif marks >= 40: return "C"
    else: return "Fail"

# ================= START MENU =================
def start_menu():
    clear_root()

    frame = tk.Frame(root, bg=BG)
    frame.pack(expand=True)

    tk.Label(frame, text="BrandeD Technologies",
             font=("Arial", 26, "bold"), bg=BG).pack(pady=20)

    tk.Button(frame, text="Admin Login",
              width=20, bg=ACCENT, fg="white",
              command=login_screen).pack(pady=10)

# ================= LOGIN =================
def login_screen():
    clear_root()

    frame = tk.Frame(root, bg=BG)
    frame.pack(expand=True)

    tk.Label(frame, text="Welcome From BrandeD",
             font=("Arial", 22, "bold"), bg=BG).pack(pady=20)

    tk.Label(frame, text="Username", bg=BG).pack()
    user = tk.Entry(frame, width=30)
    user.pack()

    tk.Label(frame, text="Password", bg=BG).pack()
    pwd = tk.Entry(frame, show="*", width=30)
    pwd.pack()

    def login():
        if user.get() == ADMIN_USERNAME and pwd.get() == ADMIN_PASSWORD:
            dashboard()
        else:
            messagebox.showerror("Error", "Invalid Login")

    tk.Button(frame, text="Login", width=20,
              bg=ACCENT, fg="white",
              command=login).pack(pady=15)

# ================= DASHBOARD =================
def dashboard():
    clear_root()

    sidebar = tk.Frame(root, bg=SIDEBAR, width=200)
    sidebar.pack(side="left", fill="y")

    content = tk.Frame(root, bg=BG)
    content.pack(side="right", fill="both", expand=True)

    def nav(text, cmd):
        tk.Button(sidebar, text=text, bg=BTN, fg="white",
                  width=25, pady=10,
                  command=lambda: cmd(content)).pack(pady=5)

    nav("Add Student", add_student)
    nav("View Students", lambda f: view_data(f, "students.xlsx"))

    nav("Add Teacher", add_teacher)
    nav("View Teachers", lambda f: view_data(f, "teachers.xlsx"))

    nav("Add Course", add_course)
    nav("View Courses", lambda f: view_data(f, "courses.xlsx"))

    tk.Button(sidebar, text="Logout", bg="red", fg="white",
              command=start_menu).pack(side="bottom", pady=20)

# ================= VIEW DATA =================
def view_data(frame, file):
    clear_frame(frame)

    if not os.path.exists(file):
        messagebox.showerror("Error", "File not found")
        return

    wb = load_workbook(file)
    ws = wb.active

    tk.Label(frame, text="Search").pack()
    search_var = tk.StringVar()
    tk.Entry(frame, textvariable=search_var).pack()

    table_frame = tk.Frame(frame)
    table_frame.pack(fill="both", expand=True)

    tree = ttk.Treeview(table_frame)
    tree.pack(fill="both", expand=True)

    headers = [cell.value for cell in ws[1]]

    tree["columns"] = ["row_id"] + headers
    tree["show"] = "headings"

    tree.heading("row_id", text="ID")
    tree.column("row_id", width=50)

    for col in headers:
        tree.heading(col, text=col)
        tree.column(col, width=120)

    data = []
    for i, row in enumerate(ws.iter_rows(min_row=2, values_only=True), start=2):
        data.append((i,) + row)

    def load_table(rows):
        tree.delete(*tree.get_children())
        for r in rows:
            tree.insert("", "end", values=r)

    load_table(data)

    def search_data(*args):
        keyword = search_var.get().lower()
        filtered = [r for r in data if any(keyword in str(c).lower() for c in r)]
        load_table(filtered)

    search_var.trace_add("write", search_data)

    def delete_row():
        selected = tree.selection()
        if not selected:
            return

        row = tree.item(selected[0])["values"]
        ws.delete_rows(row[0])
        wb.save(file)

        messagebox.showinfo("Deleted", "Record deleted")
        view_data(frame, file)

    tk.Button(frame, text="Delete Selected", bg="red", fg="white",
              command=delete_row).pack(pady=5)

    def edit_popup(event):
        selected = tree.selection()
        if not selected:
            return

        row = tree.item(selected[0])["values"]
        excel_row = row[0]

        win = tk.Toplevel(root)
        entries = []

        for i, val in enumerate(row[1:]):
            tk.Label(win, text=headers[i]).pack()
            e = tk.Entry(win)
            e.insert(0, val)
            e.pack()
            entries.append(e)

        def save():
            for j in range(len(headers)):
                ws.cell(excel_row, j+1).value = entries[j].get()

            wb.save(file)
            messagebox.showinfo("Updated", "Record updated")
            win.destroy()
            view_data(frame, file)

        tk.Button(win, text="Save", bg="green", fg="white",
                  command=save).pack()

    tree.bind("<Double-1>", edit_popup)

# ================= ADD STUDENT =================
def add_student(frame):
    clear_frame(frame)

    fields = ["Name","Age","Gender","Roll","Class","Marks"]
    entries = {}

    for f in fields:
        tk.Label(frame, text=f).pack()
        e = tk.Entry(frame)
        e.pack()
        entries[f] = e

    def save():
        file = "students.xlsx"
        create_excel(file, fields+["Grade"])

        grade = calculate_grade(entries["Marks"].get())

        wb = load_workbook(file)
        ws = wb.active
        ws.append([entries[f].get() for f in fields] + [grade])
        wb.save(file)

        messagebox.showinfo("Saved", "Student Added")

    tk.Button(frame, text="Save", bg=ACCENT, fg="white", command=save).pack()

# ================= ADD TEACHER =================
def add_teacher(frame):
    clear_frame(frame)

    fields = ["Name","Age","Gender","ID","Subject","Salary"]
    entries = {}

    for f in fields:
        tk.Label(frame, text=f).pack()
        e = tk.Entry(frame)
        e.pack()
        entries[f] = e

    def save():
        file = "teachers.xlsx"
        create_excel(file, fields)

        wb = load_workbook(file)
        ws = wb.active
        ws.append([entries[f].get() for f in fields])
        wb.save(file)

        messagebox.showinfo("Saved", "Teacher Added")

    tk.Button(frame, text="Save", bg=ACCENT, fg="white", command=save).pack()

# ================= ADD COURSE =================
def add_course(frame):
    clear_frame(frame)

    fields = ["Course Name","Code","Teacher"]
    entries = {}

    for f in fields:
        tk.Label(frame, text=f).pack()
        e = tk.Entry(frame)
        e.pack()
        entries[f] = e

    def save():
        file = "courses.xlsx"
        create_excel(file, fields)

        wb = load_workbook(file)
        ws = wb.active
        ws.append([entries[f].get() for f in fields])
        wb.save(file)

        messagebox.showinfo("Saved", "Course Added")

    tk.Button(frame, text="Save", bg=ACCENT, fg="white", command=save).pack()

# ================= START =================
start_menu()
root.mainloop()
```

```python
import customtkinter as ctk
from tkinter import ttk, messagebox
from openpyxl import Workbook, load_workbook
import os

# ========== CONFIG ==========
ctk.set_appearance_mode("light")  # "dark" or "light"
ctk.set_default_color_theme("blue")

ADMIN_USERNAME = "admin"
ADMIN_PASSWORD = "1234"

# ========== WINDOW ==========
root = ctk.CTk()
root.title("BrandeD Technologies")
root.geometry("1000x600")

# ========== HELPERS ==========
def clear_root():
    for w in root.winfo_children():
        w.destroy()

def clear_frame(frame):
    for w in frame.winfo_children():
        w.destroy()

def create_excel(file, headers):
    if not os.path.exists(file):
        wb = Workbook()
        ws = wb.active
        ws.append(headers)
        wb.save(file)

# ========== START ==========
def start_menu():
    clear_root()

    frame = ctk.CTkFrame(root)
    frame.pack(expand=True)

    ctk.CTkLabel(frame, text="BrandeD Technologies",
                 font=("Arial", 28, "bold")).pack(pady=20)

    ctk.CTkButton(frame, text="BrandeD Login",
                  width=200, command=login_screen).pack(pady=10)

# ========== LOGIN ==========
def login_screen():
    clear_root()

    frame = ctk.CTkFrame(root)
    frame.pack(expand=True)

    ctk.CTkLabel(frame, text="BrandeD login",
                 font=("Arial", 24, "bold")).pack(pady=20)

    user = ctk.CTkEntry(frame, placeholder_text="Username", width=250)
    user.pack(pady=5)

    pwd = ctk.CTkEntry(frame, placeholder_text="Password", show="*", width=250)
    pwd.pack(pady=5)

    def login():
        if user.get() == ADMIN_USERNAME and pwd.get() == ADMIN_PASSWORD:
            dashboard()
        else:
            messagebox.showerror("Error", "Invalid Login")

    ctk.CTkButton(frame, text="Login", width=200,
                  command=login).pack(pady=15)

# ========== DASHBOARD ==========
def dashboard():
    clear_root()

    sidebar = ctk.CTkFrame(root, width=200)
    sidebar.pack(side="left", fill="y")

    content = ctk.CTkFrame(root)
    content.pack(side="right", fill="both", expand=True)

    def nav(text, cmd):
        ctk.CTkButton(sidebar, text=text, width=180,
                      command=lambda: cmd(content)).pack(pady=5)

    nav("Add Student", add_student)
    nav("View Students", lambda f: view_data(f, "students.xlsx"))

    nav("Add Teacher", add_teacher)
    nav("View Teachers", lambda f: view_data(f, "teachers.xlsx"))

    nav("Add Course", add_course)
    nav("View Courses", lambda f: view_data(f, "courses.xlsx"))

    ctk.CTkButton(sidebar, text="Logout", fg_color="red",
                  command=start_menu).pack(side="bottom", pady=20)

# ========== VIEW DATA ==========
def view_data(frame, file):
    clear_frame(frame)

    if not os.path.exists(file):
        messagebox.showerror("Error", "No file found")
        return

    wb = load_workbook(file)
    ws = wb.active

    search_var = ctk.StringVar()
    ctk.CTkEntry(frame, placeholder_text="Search...",
                 textvariable=search_var).pack(pady=5)

    table_frame = ctk.CTkFrame(frame)
    table_frame.pack(fill="both", expand=True)

    tree = ttk.Treeview(table_frame)
    tree.pack(fill="both", expand=True)

    headers = [cell.value for cell in ws[1]]

    tree["columns"] = ["id"] + headers
    tree["show"] = "headings"

    tree.heading("id", text="ID")
    tree.column("id", width=50)

    for col in headers:
        tree.heading(col, text=col)
        tree.column(col, width=120)

    data = []
    for i, row in enumerate(ws.iter_rows(min_row=2, values_only=True), start=2):
        data.append((i,) + row)

    def load(rows):
        tree.delete(*tree.get_children())
        for r in rows:
            tree.insert("", "end", values=r)

    load(data)

    def search(*args):
        keyword = search_var.get().lower()
        filtered = [r for r in data if any(keyword in str(c).lower() for c in r)]
        load(filtered)

    search_var.trace_add("write", search)

    def delete():
        selected = tree.selection()
        if not selected:
            return

        row = tree.item(selected[0])["values"]
        ws.delete_rows(row[0])
        wb.save(file)

        messagebox.showinfo("Deleted", "Record deleted")
        view_data(frame, file)

    ctk.CTkButton(frame, text="Delete Selected",
                  fg_color="red", command=delete).pack(pady=5)

    def edit(event):
        selected = tree.selection()
        if not selected:
            return

        row = tree.item(selected[0])["values"]
        excel_row = row[0]

        win = ctk.CTkToplevel(root)
        entries = []

        for i, val in enumerate(row[1:]):
            ctk.CTkLabel(win, text=headers[i]).pack()
            e = ctk.CTkEntry(win)
            e.insert(0, val)
            e.pack()
            entries.append(e)

        def save():
            for j in range(len(headers)):
                ws.cell(excel_row, j+1).value = entries[j].get()

            wb.save(file)
            messagebox.showinfo("Updated", "Record updated")
            win.destroy()
            view_data(frame, file)

        ctk.CTkButton(win, text="Save", command=save).pack(pady=10)

    tree.bind("<Double-1>", edit)

# ========== ADD FORMS ==========
def add_student(frame):
    clear_frame(frame)

    fields = ["Name","Age","Gender","Roll","Class","Marks"]
    entries = {}

    for f in fields:
        ctk.CTkLabel(frame, text=f).pack()
        e = ctk.CTkEntry(frame)
        e.pack()
        entries[f] = e

    def save():
        file = "students.xlsx"
        create_excel(file, fields)

        wb = load_workbook(file)
        ws = wb.active
        ws.append([entries[f].get() for f in fields])
        wb.save(file)

        messagebox.showinfo("Saved", "Student Added")

    ctk.CTkButton(frame, text="Save", command=save).pack(pady=10)

def add_teacher(frame):
    clear_frame(frame)

    fields = ["Name","Age","Gender","ID","Subject","Salary"]
    entries = {}

    for f in fields:
        ctk.CTkLabel(frame, text=f).pack()
        e = ctk.CTkEntry(frame)
        e.pack()
        entries[f] = e

    def save():
        file = "teachers.xlsx"
        create_excel(file, fields)

        wb = load_workbook(file)
        ws = wb.active
        ws.append([entries[f].get() for f in fields])
        wb.save(file)

        messagebox.showinfo("Saved", "Teacher Added")

    ctk.CTkButton(frame, text="Save", command=save).pack(pady=10)

def add_course(frame):
    clear_frame(frame)

    fields = ["Course","Code","Teacher"]
    entries = {}

    for f in fields:
        ctk.CTkLabel(frame, text=f).pack()
        e = ctk.CTkEntry(frame)
        e.pack()
        entries[f] = e

    def save():
        file = "courses.xlsx"
        create_excel(file, fields)

        wb = load_workbook(file)
        ws = wb.active
        ws.append([entries[f].get() for f in fields])
        wb.save(file)

        messagebox.showinfo("Saved", "Course Added")

    ctk.CTkButton(frame, text="Save", command=save).pack(pady=10)

# ========== START ==========
start_menu()
root.mainloop()
```
