```python
from openpyxl import Workbook, load_workbook
import os

# ================== PARENT CLASS ==================
class Person:
    def __init__(self, name="", age="", gender=""):
        self.name = name
        self.age = age
        self.gender = gender


# ================== STUDENT CLASS ==================
class Student(Person):
    def __init__(self):
        super().__init__()

    def add_student(self):
        self.name = input("Enter Name: ")
        self.age = input("Enter Age: ")
        self.gender = input("Enter Gender: ")
        self.roll_no = input("Enter Roll Number: ")
        self.class_name = input("Enter Class Name: ")
        self.marks = input("Enter Marks: ")

    def calculate_grade(self):
        marks = int(self.marks)
        if marks >= 80: return "A"
        elif marks >= 60: return "B"
        elif marks >= 40: return "C"
        else: return "Fail"

    def save_to_excel(self):
        file = "students_data.xlsx"

        if not os.path.exists(file):
            wb = Workbook()
            wb.active.append(["Name","Age","Gender","Roll No","Class","Marks","Grade"])
            wb.save(file)

        wb = load_workbook(file)
        sheet = wb.active
        sheet.append([self.name,self.age,self.gender,self.roll_no,self.class_name,self.marks,self.calculate_grade()])
        wb.save(file)
        print("✅ Student saved!")

    def show_students(self):
        file = "students_data.xlsx"
        if not os.path.exists(file):
            print("No data!")
            return

        wb = load_workbook(file)
        for row in wb.active.iter_rows(values_only=True):
            print(row)

    def delete_student(self):
        file = "students_data.xlsx"
        roll = input("Enter Roll No: ")

        wb = load_workbook(file)
        sheet = wb.active

        for i in range(2, sheet.max_row+1):
            if str(sheet.cell(i,4).value) == roll:
                sheet.delete_rows(i)
                wb.save(file)
                print("Deleted")
                return
        print("Not found")


# ================== TEACHER CLASS ==================
class Teacher(Person):
    def __init__(self):
        super().__init__()

    def add_teacher(self):
        self.name = input("Enter Name: ")
        self.age = input("Enter Age: ")
        self.gender = input("Enter Gender: ")
        self.id = input("Enter Teacher ID: ")
        self.subject = input("Enter Subject: ")
        self.salary = input("Enter Salary: ")

    def save_teacher(self):
        file = "teachers_data.xlsx"

        if not os.path.exists(file):
            wb = Workbook()
            wb.active.append(["Name","Age","Gender","ID","Subject","Salary"])
            wb.save(file)

        wb = load_workbook(file)
        wb.active.append([self.name,self.age,self.gender,self.id,self.subject,self.salary])
        wb.save(file)
        print("✅ Teacher saved!")

    def show_teachers(self):
        file = "teachers_data.xlsx"
        if not os.path.exists(file):
            print("No data!")
            return

        wb = load_workbook(file)
        for row in wb.active.iter_rows(values_only=True):
            print(row)

    def delete_teacher(self):
        file = "teachers_data.xlsx"
        tid = input("Enter ID: ")

        wb = load_workbook(file)
        sheet = wb.active

        for i in range(2, sheet.max_row+1):
            if str(sheet.cell(i,4).value) == tid:
                sheet.delete_rows(i)
                wb.save(file)
                print("Deleted")
                return
        print("Not found")


# ================== COURSE CLASS ==================
class Course:
    file = "courses.xlsx"

    @staticmethod
    def init():
        if not os.path.exists(Course.file):
            wb = Workbook()
            wb.active.append(["Course Name","Course Code","Teacher"])
            wb.save(Course.file)

    @staticmethod
    def add_course():
        Course.init()
        name = input("Course Name: ")
        code = input("Course Code: ")
        teacher = input("Teacher Name: ")

        wb = load_workbook(Course.file)
        wb.active.append([name,code,teacher])
        wb.save(Course.file)
        print("✅ Course added")

    @staticmethod
    def show_courses():
        Course.init()
        wb = load_workbook(Course.file)
        for row in wb.active.iter_rows(values_only=True):
            print(row)

    @staticmethod
    def delete_course():
        Course.init()
        code = input("Enter Course Code: ")

        wb = load_workbook(Course.file)
        sheet = wb.active

        for i in range(2, sheet.max_row+1):
            if str(sheet.cell(i,2).value) == code:
                sheet.delete_rows(i)
                wb.save(Course.file)
                print("Deleted")
                return
        print("Not found")


# ================== MAIN MENU ==================
student = Student()
teacher = Teacher()

while True:
    print("\n===== MAIN MENU =====")
    print("1. Add Student")
    print("2. Show Students")
    print("3. Delete Student")
    print("4. Add Teacher")
    print("5. Show Teachers")
    print("6. Delete Teacher")
    print("7. Add Course")
    print("8. Show Courses")
    print("9. Delete Course")
    print("10. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        student.add_student(); student.save_to_excel()
    elif choice == "2":
        student.show_students()
    elif choice == "3":
        student.delete_student()
    elif choice == "4":
        teacher.add_teacher(); teacher.save_teacher()
    elif choice == "5":
        teacher.show_teachers()
    elif choice == "6":
        teacher.delete_teacher()
    elif choice == "7":
        Course.add_course()
    elif choice == "8":
        Course.show_courses()
    elif choice == "9":
        Course.delete_course()
    elif choice == "10":
        break
    else:
        print("Invalid choice")
```

```python
import tkinter as tk
from tkinter import messagebox
from openpyxl import Workbook, load_workbook
import os

# ---------- Excel Helper ----------
def create_excel(file, headers):
    if not os.path.exists(file):
        wb = Workbook()
        ws = wb.active
        ws.append(headers)
        wb.save(file)

def calculate_grade(marks):
    marks = int(marks)
    if marks >= 80:
        return "A"
    elif marks >= 60:
        return "B"
    elif marks >= 40:
        return "C"
    else:
        return "Fail"

# ---------- Main Window ----------
root = tk.Tk()
root.title("School Management System")
root.geometry("450x500")

title = tk.Label(root, text="School Management GUI", font=("Arial", 18, "bold"))
title.pack(pady=15)

# ---------- Clear Window ----------
def clear_window():
    for widget in root.winfo_children():
        widget.destroy()

# ---------- Back Button ----------
def back_button():
    tk.Button(root, text="Back", command=main_menu, width=20).pack(pady=10)

# ---------- Add Student ----------
def add_student_form():
    clear_window()

    tk.Label(root, text="Add Student", font=("Arial", 16, "bold")).pack(pady=10)

    fields = ["Name", "Age", "Gender", "Roll No", "Class", "Marks"]
    entries = {}

    for field in fields:
        tk.Label(root, text=field).pack()
        entry = tk.Entry(root, width=35)
        entry.pack()
        entries[field] = entry

    def submit():
        file = "students_data.xlsx"
        create_excel(file, ["Name", "Age", "Gender", "Roll No", "Class", "Marks", "Grade"])

        grade = calculate_grade(entries["Marks"].get())

        wb = load_workbook(file)
        ws = wb.active
        ws.append([
            entries["Name"].get(),
            entries["Age"].get(),
            entries["Gender"].get(),
            entries["Roll No"].get(),
            entries["Class"].get(),
            entries["Marks"].get(),
            grade
        ])
        wb.save(file)

        messagebox.showinfo("Success", "Student saved successfully!")

    tk.Button(root, text="Submit", command=submit, bg="green", fg="white", width=20).pack(pady=10)
    back_button()

# ---------- Delete Student ----------
def delete_student_form():
    clear_window()

    tk.Label(root, text="Delete Student", font=("Arial", 16, "bold")).pack(pady=10)

    tk.Label(root, text="Enter Roll No").pack()
    roll_entry = tk.Entry(root, width=35)
    roll_entry.pack()

    def delete_student():
        file = "students_data.xlsx"

        if not os.path.exists(file):
            messagebox.showerror("Error", "Student file not found!")
            return

        wb = load_workbook(file)
        ws = wb.active
        roll = roll_entry.get()

        for i in range(2, ws.max_row + 1):
            if str(ws.cell(i, 4).value) == roll:
                ws.delete_rows(i)
                wb.save(file)
                messagebox.showinfo("Deleted", "Student deleted successfully!")
                return

        messagebox.showerror("Not Found", "Student not found!")

    tk.Button(root, text="Delete", command=delete_student, bg="red", fg="white", width=20).pack(pady=10)
    back_button()

# ---------- Add Teacher ----------
def add_teacher_form():
    clear_window()

    tk.Label(root, text="Add Teacher", font=("Arial", 16, "bold")).pack(pady=10)

    fields = ["Name", "Age", "Gender", "Teacher ID", "Subject", "Salary"]
    entries = {}

    for field in fields:
        tk.Label(root, text=field).pack()
        entry = tk.Entry(root, width=35)
        entry.pack()
        entries[field] = entry

    def submit():
        file = "teachers_data.xlsx"
        create_excel(file, ["Name", "Age", "Gender", "ID", "Subject", "Salary"])

        wb = load_workbook(file)
        ws = wb.active
        ws.append([
            entries["Name"].get(),
            entries["Age"].get(),
            entries["Gender"].get(),
            entries["Teacher ID"].get(),
            entries["Subject"].get(),
            entries["Salary"].get()
        ])
        wb.save(file)

        messagebox.showinfo("Success", "Teacher saved successfully!")

    tk.Button(root, text="Submit", command=submit, bg="green", fg="white", width=20).pack(pady=10)
    back_button()

# ---------- Delete Teacher ----------
def delete_teacher_form():
    clear_window()

    tk.Label(root, text="Delete Teacher", font=("Arial", 16, "bold")).pack(pady=10)

    tk.Label(root, text="Enter Teacher ID").pack()
    id_entry = tk.Entry(root, width=35)
    id_entry.pack()

    def delete_teacher():
        file = "teachers_data.xlsx"

        if not os.path.exists(file):
            messagebox.showerror("Error", "Teacher file not found!")
            return

        wb = load_workbook(file)
        ws = wb.active
        teacher_id = id_entry.get()

        for i in range(2, ws.max_row + 1):
            if str(ws.cell(i, 4).value) == teacher_id:
                ws.delete_rows(i)
                wb.save(file)
                messagebox.showinfo("Deleted", "Teacher deleted successfully!")
                return

        messagebox.showerror("Not Found", "Teacher not found!")

    tk.Button(root, text="Delete", command=delete_teacher, bg="red", fg="white", width=20).pack(pady=10)
    back_button()

# ---------- Add Course ----------
def add_course_form():
    clear_window()

    tk.Label(root, text="Add Course", font=("Arial", 16, "bold")).pack(pady=10)

    fields = ["Course Name", "Course Code", "Teacher"]
    entries = {}

    for field in fields:
        tk.Label(root, text=field).pack()
        entry = tk.Entry(root, width=35)
        entry.pack()
        entries[field] = entry

    def submit():
        file = "courses.xlsx"
        create_excel(file, ["Course Name", "Course Code", "Teacher"])

        wb = load_workbook(file)
        ws = wb.active
        ws.append([
            entries["Course Name"].get(),
            entries["Course Code"].get(),
            entries["Teacher"].get()
        ])
        wb.save(file)

        messagebox.showinfo("Success", "Course saved successfully!")

    tk.Button(root, text="Submit", command=submit, bg="green", fg="white", width=20).pack(pady=10)
    back_button()

# ---------- Main Menu ----------
def main_menu():
    clear_window()

    tk.Label(root, text="School Management GUI", font=("Arial", 18, "bold")).pack(pady=20)

    tk.Button(root, text="Add Student", command=add_student_form, width=25).pack(pady=5)
    tk.Button(root, text="Delete Student", command=delete_student_form, width=25).pack(pady=5)

    tk.Button(root, text="Add Teacher", command=add_teacher_form, width=25).pack(pady=5)
    tk.Button(root, text="Delete Teacher", command=delete_teacher_form, width=25).pack(pady=5)

    tk.Button(root, text="Add Course", command=add_course_form, width=25).pack(pady=5)

# Start GUI
main_menu()
root.mainloop()
```
