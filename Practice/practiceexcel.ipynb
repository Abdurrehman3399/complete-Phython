
## 1. Create Excel File and Write Data

```python
from openpyxl import Workbook

# Create workbook
wb = Workbook()

# Select sheet
sheet = wb.active

# Write data
sheet["A1"] = "Name"
sheet["B1"] = "Age"

sheet["A2"] = "Ali"
sheet["B2"] = 20

sheet["A3"] = "Ahmed"
sheet["B3"] = 22

# Save file
wb.save("students.xlsx")

print("Excel file created successfully")
```

```python
from openpyxl import load_workbook

# Load excel file
wb = load_workbook("students.xlsx")

# Select sheet
sheet = wb.active

# Read data
print(sheet["A2"].value)
print(sheet["B2"].value)
```

```python
from openpyxl import load_workbook

wb = load_workbook("students.xlsx")
sheet = wb.active

for row in sheet.iter_rows(values_only=True):
    print(row)
```

```python
from openpyxl import load_workbook

wb = load_workbook("students.xlsx")
sheet = wb.active

sheet.append(["Usman", 25])

wb.save("students.xlsx")

print("New data added")
```

```python
from openpyxl import Workbook
from openpyxl.worksheet.table import Table, TableStyleInfo

# Create a new workbook
wb = Workbook()
ws = wb.active
ws.title = "Student"

# Add headings
data = [
    ["ID", "Name", "Age", "Course"],
    [1, "Usman", 25, "Python"],
    [2, "Ali", 22, "Data Science"],
    [3, "Ahmed", 24, "AI"],
    [4, "Sara", 23, "Web Development"]
]

# Insert data into sheet
for row in data:
    ws.append(row)

# Create a table
table = Table(displayName="StudentTable", ref="A1:D5")

# Add style to table
style = TableStyleInfo(
    name="TableStyleMedium9",
    showFirstColumn=False,
    showLastColumn=False,
    showRowStripes=True,
    showColumnStripes=True
)
table.tableStyleInfo = style

# Add table to worksheet
ws.add_table(table)

# Save file
wb.save("students.xlsx")

print("Excel file 'students.xlsx' created successfully!")
```
