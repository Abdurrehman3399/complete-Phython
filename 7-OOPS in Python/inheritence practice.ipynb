Question 1: Animal and Dog

Create a class Animal with:

attribute: name

method: sound()


Create another class Dog that inherits from Animal. Add:

attribute: breed

method: display()


Task

Print:

Dog name

Dog breed

Dog sound

```python
class Animal:
  def __init__(self, name):
    self.name = name

  def sound(self):
    return "Woof!"

class Dog(Animal):
  def __init__(self, name, breed):
    super().__init__(name)
    self.breed = breed



  def Display(self):
    print("Dog Name:", self.name)
    print("Dog Breed:", self.breed)
    print("Dog Sound:", self.sound())


Dog1 = Dog("Max","")
Dog1.Display()
```

Question 2: Person and Student

Create a class Person with:

name

age

Create a class Student that inherits from Person. Add:

roll_number

Task

Create a method to display:

Name

Age

Roll Number

```python
class person :
  def __init__(self,name,age):
    self.name = name
    self.age = age

class Student(person):
  def __init__(self,name,age,roll_number):
    person.__init__(self,name, age)
    self.roll_number = roll_number

  def Display(self):
    print("Name:", self.name)
    print("Age:", self.age)
    print("Roll Number:", self.roll_number)


p1 = Student("Admin", "12", "123")
p1.Display()
```

Question 3: Vehicle and Car

Create a class Vehicle with:

company


Create a class Car that inherits from Vehicle. Add:

model


Task

Display:

Company name

Car model

```python
class Vehicle:
  def __init__(self, company):
    self.company = company

class Car(Vehicle):
  def __init__(self, company, model):
    super().__init__(company)
    self.model = model

  def Display(self):
    print("Company:", self.company)
    print("Model:", self.model)


p1 = Car("Toyota", "Mark X")
p1.Display()
```


Question 4: Teacher and MathTeacher

Create a class Teacher with:

teacher_name


Create another class MathTeacher that inherits from Teacher. Add:

subject


Task

Print:

Teacher name

Subject name






```python
class Teacher:
  def __init__(self, teacher_name):
    self.name = teacher_name

class Math_teacher(Teacher):
  def __init__(self, teacher_name, sub ):
    super().__init__(teacher_name)
    self.sub = sub

  def Display(self):
    print("teacher_name:", self.name)
    print("sub:", self.sub)

p1 = Math_teacher ("Rahimullah", "Maths")
p1.Display()
```
