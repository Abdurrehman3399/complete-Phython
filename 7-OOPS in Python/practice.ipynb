ask: Create a Python program using a class named Student to store and manage student information for a Class 8 level system.
The class should include student details such as name, roll number, age, class, and marks out of 100.


```python
class Student:
    def __init__(self):
        self.name = "" 
        self.age = 0
        self.rollno = 0
        self.marks =0
        self.classname = ""
    def inputdata (self):
        self.name = (input("Enter your name :"))
        self.age =  int (input ("Enter your age :"))
        self.rollno =  int (input("Enter your roll no :"))
        self.marks = int (input ("Enter your Marks :"))
        self.classname = input ("Enter your ClassName :")

    def grade (self):
        if self.marks >= 90:
            return ("A+")
        elif self.marks >= 80:
            return ("A")
        elif self.marks >= 70:
            return ("B")
        elif self.marks >= 60:
            return ("c")
        else:
            return("Failed")
        
    def display (self):
            print("name", self.name)
            print("age" ,self.age)
            print("marks" ,self.marks)
            print("rollno" ,self.age)
            print("classname" , self.classname)      




        
stu1 = Student()
stu1.inputdata()
stu1.display()
stu1.grade()



   
```
